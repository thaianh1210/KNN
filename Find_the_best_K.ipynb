#How to select K in KNN


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_path = '/content/Iris (2).csv'
name = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

dataset = pd.read_csv('/content/Iris (2).csv')
dataset
```

```python
X = dataset.iloc[:, 1:5].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
```

```python
scale = StandardScaler()
X_train_scale = scale.fit_transform(X_train)
X_test_scale = scale.transform(X_test)
```

```python
knn_cls = KNeighborsClassifier(n_neighbors= 5)
knn_cls.fit(X_train, y_train)
y_pred = knn_cls.predict(X_test)
```

```python
y_pred != y_test
```

```python
print(classification_report(y_pred, y_test))
```

```python
#K selection
k_values = range(1, 31)
accuracies = []
def KNN(X_train, y_train, X_test, y_test):
  for i in k_values:
    knn_cls = KNeighborsClassifier(n_neighbors= i)
    knn_cls.fit(X_train, y_train)
    y_pred = knn_cls.predict(X_test)
    accuracy = accuracy_score(y_pred, y_test)
    accuracies.append(accuracy)

KNN(X_train_scale, y_train, X_test_scale, y_test)
```

```python
def plot_result(accuracies, k_values):
  plt.figure(figsize = (10, 8))
  plt.plot(k_values, accuracies, color='blue', marker='o',
        markerfacecolor='yellow', markersize=10)
  plt.title('Accuracy vs K Value')
  plt.xlabel('K Values')
  plt.ylabel('Accuracy')
  plt.legend()
  plt.show()
```

```python
plot_result(accuracies, k_values)
```
